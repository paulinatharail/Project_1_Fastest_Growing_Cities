```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("merge_city_pop_greater_than_200k.csv")
```

```python
df_top = df.head(10)[["City, State", 
                      "5-Year Average Population Growth Rate",
                      "5-Year Average Median Household Income Growth Rate", 
                      "5-Year Average Employed Growth Rate",
                      "5-Year Average Monthly Rent Growth Rate",
                      "5-Year Average Owner Monthly Costs Growth Rate"]]

df_top = df_top.rename(columns={"5-Year Average Population Growth Rate": "Avg. Population\nGrowth Rate(%)",
                       "5-Year Average Median Household Income Growth Rate": "Avg. Median\nGrowth Rate(%)",
                       "5-Year Average Employed Growth Rate": "Avg. Employed\nGrowth Rate(%)",
                       "5-Year Average Monthly Rent Growth Rate": "Avg. Monthly Rent\nGrowth Rate(%)",
                       "5-Year Average Owner Monthly Costs Growth Rate": "Avg. Owner Monthly Cost\nGrowth Rate(%)"
                      })
df_top
```

```python
df.columns
```

```python
top_10_cities_list=[]

for i in range(10):
    data = df.iloc[i]
    city1=data['City, State']
    top_10_cities_list.append(city1)
```

```python
plt.figure(figsize=(30,20))
year_list_2 = ['2012', '2013', '2014', '2015', '2016']
j=1
print(top_10_cities_list)
for city in top_10_cities_list:
   
    print(f"#### Processing {city} ####")
    
    s2=df.loc[df["City, State"]==city, :]
    #print(type(s2))

    P_list=[
        s2["2012 Total Population"].iloc[0],
        s2["2013 Total Population"].iloc[0],
        s2["2014 Total Population"].iloc[0],
        s2["2015 Total Population"].iloc[0],
        s2["2016 Total Population"].iloc[0]
    ]
    I_list=[
        s2["2012 Median Household Income"].iloc[0], 
        s2["2013 Median Household Income"].iloc[0],   
        s2["2014 Median Household Income"].iloc[0], 
        s2["2015 Median Household Income"].iloc[0],
        s2["2016 Median Household Income"].iloc[0]
    ]
    E_list=[
        s2["2012 Employed"].iloc[0], 
        s2["2013 Employed"].iloc[0], 
        s2["2014 Employed"].iloc[0], 
        s2["2015 Employed"].iloc[0],
        s2["2016 Employed"].iloc[0]
    ]
    R_list=[
        s2["2012 Monthly Rent"].iloc[0], 
        s2["2013 Monthly Rent"].iloc[0], 
        s2["2014 Monthly Rent"].iloc[0], 
        s2["2015 Monthly Rent"].iloc[0],
        s2["2016 Monthly Rent"].iloc[0]
    ]
    
    C_list=[
        s2["2012 Owner Monthly Costs"].iloc[0], 
        s2["2013 Owner Monthly Costs"].iloc[0], 
        s2["2014 Owner Monthly Costs"].iloc[0], 
        s2["2015 Owner Monthly Costs"].iloc[0],
        s2["2016 Owner Monthly Costs"].iloc[0]
    ]
    
        
    plt.subplot(2,5 ,j)
    plt.plot(year_list_2, P_list, color= "red", linewidth=4, label="Population")
    plt.plot(year_list_2, I_list, color="blue", linewidth=4, label="Household Income" )
    plt.plot(year_list_2, E_list, color="green", linewidth=4, label="Employment" )
    plt.plot(year_list_2, R_list, color="orange", linewidth=4, label="Rent")
    plt.plot(year_list_2, C_list, color="darkgrey", linewidth=4, label="Owner Monthly Cost")

    j=j+1
    
    plt.title(city, fontsize=28)
    
    x_ticks = year_list_2
    plt.tick_params(axis='x', labelsize=20)
    plt.tick_params(axis='y', labelsize=14)
    
plt.legend(['Population', 'Household Income', 'Employed', 'Monthly Rent', 'Owner Monthly Cost'], fontsize=24, loc='center left', bbox_to_anchor=(1.12,1.12))

plt.savefig("Images/speed_of_change.png")
```
