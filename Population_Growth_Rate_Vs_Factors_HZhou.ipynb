```python
import os
import csv
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt

df = pd.read_csv("merge_city_pop_greater_than_200k.csv")
#df.head()
```

```python
### rank cities #######

#df_sorted=df.sort_values("5-Year Average Population Growth Rate",ascending=False)
#print(df_sorted.head())

top_10_cities_list=[]

for i in range(10):
    data = df.iloc[i]
    city1=data['City, State']
    top_10_cities_list.append(city1)
print(f"Here are the top 10 cities:  {top_10_cities_list}")
#df_sorted.to_csv('df_sorted.csv')

years = ['2012 to 2013', '2013 to 2014', '2014 to 2015', '2015 to 2016']

plt.figure(figsize=(30,20))
i=1
print(top_10_cities_list)
for city in top_10_cities_list:
   
    print(f"#### Processing {city} ####")
    
    s=df.loc[df["City, State"]==city, :]
    # print("### printing s ####")
    # print(s)
    P_list=[
        s["Population Growth Rate (2012 to 2013)"].iloc[0],
        s["Population Growth Rate (2013 to 2014)"].iloc[0],
        s["Population Growth Rate (2014 to 2015)"].iloc[0],
        s["Population Growth Rate (2015 to 2016)"].iloc[0]
    ]
    I_list=[
        s["Median Household Income Growth Rate (2012 to 2013)"].iloc[0], 
        s["Median Household Income Growth Rate (2013 to 2014)"].iloc[0], 
        s["Median Household Income Growth Rate (2014 to 2015)"].iloc[0], 
        s["Median Household Income Growth Rate (2015 to 2016)"].iloc[0]
    ]
    E_list=[
        s["Employed Growth Rate (2012 to 2013)"].iloc[0], 
        s["Employed Growth Rate (2013 to 2014)"].iloc[0], 
        s["Employed Growth Rate (2014 to 2015)"].iloc[0], 
        s["Employed Growth Rate (2015 to 2016)"].iloc[0]
    ]
    R_list=[
        s["Monthly Rent Growth Rate (2012 to 2013)"].iloc[0], 
        s["Monthly Rent Growth Rate (2013 to 2014)"].iloc[0], 
        s["Monthly Rent Growth Rate (2014 to 2015)"].iloc[0], 
        s["Monthly Rent Growth Rate (2015 to 2016)"].iloc[0]
    ]
    #print("### printing s[] ###")
    #print(s["Median Household Income Growth Rate (2012 to 2013)"])
    
    plt.subplot(2,5 ,i)
    plt.plot(years, P_list, color= "red", label="Population growth rate")
    plt.plot(years, I_list, color="blue", label="Income growth rate" )
    plt.plot(years, E_list, color="green", label="Employment growth rate" )
    plt.plot(years, R_list, color="gold", label="Rent growth rate")
    
    
    # print(f"####Finish prcessing {city} #####")  
    

    i=i+1
    
```
