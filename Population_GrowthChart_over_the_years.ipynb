```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import merge_city_pop_greater_than_200k.csv
pop_200_df = pd.read_csv("merge_city_pop_greater_than_200k.csv")
pop_200_df.head()
```

```python
pop_200_df["2012 Median Household Income"] = pop_200_df["2012 Median Household Income"].map("${:,.2f}".format)
    
pop_200_df.head()
```

```python
#display columns
print (pop_200_df.columns)
```

```python



#2012 income vs population
plt.scatter(pop_200_df['2012 Median Household Income'], pop_200_df['2012 Total Population']/10000
           , marker="o", facecolors="palevioletred")


#2013 income vs population
plt.scatter(pop_200_df['2013 Median Household Income'], pop_200_df['2013 Total Population']/10000
           , marker="^", facecolors="gold")

#2014 income vs population
plt.scatter(pop_200_df['2014 Median Household Income'], pop_200_df['2014 Total Population']/10000
           , marker="s", facecolors="chartreuse")

#2015 income vs population
plt.scatter(pop_200_df['2015 Median Household Income'], pop_200_df['2015 Total Population']/10000
           , marker="p", facecolors="royalblue")

#2016 income vs population
plt.scatter(pop_200_df['2016 Median Household Income'], pop_200_df['2016 Total Population']/10000
           , marker="d", facecolors="plum")


#legend
plt.legend(['2012','2013','2014','2015','2016'], loc="best")

#Chart title
plt.title("Population Vs Income (2012-2016)")

#Chart axes
plt.xlabel("Median Household Income")
plt.ylabel("Population (10,000s)")

#gridlines
plt.grid(axis="both")

#display graph
plt.show()

plt.savefig("Images/Population_Vs_Income.png")
```
